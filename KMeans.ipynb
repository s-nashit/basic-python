```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('salary.csv')
```

```python
df.sample()
```

```python
plt.scatter(df['age'], df['salary'])
```

```python
model = KMeans(n_clusters = 2)
```

```python
df['cluster'] = model.fit_predict(df[['age', 'salary']])
```

```python
model.cluster_centers_
```

```python
model.cluster_centers_[:,0]
```

```python
model.cluster_centers_[:,1]
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
```

```python
plt.scatter(df1['age'], df1['salary'], marker='*', c='c')
plt.scatter(df2['age'], df2['salary'], marker='>', c='m')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1], marker='D', color='k')
```
