```python
import mpl_toolkits as mpt
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5,6,7])
y = np.array([3, 5,7,8,9,3,9])
z = np.ones(7)
```

```python
xsize=np.ones(7)
ysize=np.ones(7)
zsize=np.array([2,3,2,5,2,4,2])
```

```python
plt.figure(figsize=(4,4))
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.bar3d(x,y,z,xsize,ysize,zsize)
plt.show()
```
