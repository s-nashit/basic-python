```python
#Catplot, Pairplot, Facet Grid
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df =sns.load_dataset('tips')
```

```python
df.head(1)
```

```python
figr = sns.FacetGrid(df, col='sex', hue='day', palette='cool')
figr.map(plt.scatter, 'total_bill','tip').add_legend()
plt.show()
```

```python
#figr.map(plt.bar, 'total_bill','tip')
```

```python
#PAIR PLOT
df.head(1)
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue='sex', hue_order=['Female','Male'], palette='Accent', kind='scatter', markers=['D','<'])
#kind : hist, scatter, kde, reg
plt.show()
```

```python
# If the group size is large, the bill value and tip also increases
```

```python
#CAT PLOT
#plt.figure(figsize=(3,3))
sns.catplot(x='size',y ='tip', data = df.head(30), hue='sex', palette='Accent',kind='point', height=3)
#Scatters -> default, swarm, strip
#Distributions-> box, boxen,violin
#estimatation -> point,bar,count
plt.show()
```

```python
#Point plot
sns.pointplot(x='size', y='tip', data=df, hue='sex')
```
