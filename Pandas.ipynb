```python
y2 = ['Damodar', 'Mukesh', 'Shrey', 104]
a2 = pd.Series(y2, index= [101,102,103,104])
a2
```

```python
y1 = {'name':['Damodar', 'Mukesh', 'Shrey', 'Smriti']}
a = pd.Series(y1)
print(a)
type(a)
```

```python
y2 = ('Damodar', 'Mukesh', 'Shrey', 104)
a2 = pd.Series(y2)
print(a2, type(a2))
```

```python
y3 = {'name': ['Dam', 'Shr', 'Muk'], 'age':[21, 21,20]}
a3 = pd.Series(y3)
print(a3)
print(type(a3))
```

```python
y4 = pd.Series(10, index=[101, 102, 103, 104, 105])
y5 = pd.Series(20, index=[101,102,103])
print(y4)
print(y5)
```

```python
b = (y4*y5)
c = pd.Series(b, dtype='int64')
c
```
