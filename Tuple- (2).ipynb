```python
#Tuple
```

```python
tuple = ("apple", "banana", "cherry", "apple", "cherry")
print(tuple)
```

```python
tuple = ("apple", "banana", "cherry")
print(tuple[0])  # Access 1st item
print(tuple[1])  # Access 2nd item
```

```python
tuple = ("apple", "banana", "cherry")
print(tuple[-1])  #Negative Indexing - Access the last item
```

```python
tuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(tuple[2:5])  
# 0 being the first index point; this range refers to the item 3rd to 6th 
```

```python
tuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(tuple[:4])   # returns upto 4th item
```

```python
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:]) #prints items 3rd & later; 0 being the 1st
```

```python
x = ("apple", "banana", "cherry")
y = list(x)  #Convert tuple into list
print(y)
```

```python
thistuple = ("apple", "banana", "cherry")
thistuple.append("orange") # This will raise an error
print(thistuple)
```

```python
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
print(thistuple)
```

```python
fruits = ("apple", "banana", "cherry")
print(fruits)
```

```python
fruits = ("apple", "banana", "cherry")
(green, yellow, red) = fruits
print(green)   #Unpacking
print(yellow)
print(red)
```

```python
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")
(green, yellow, *red) = fruits   # Using Asterisk*
print(green)
print(yellow)
print(red)
```

```python
fruits = ("apple", "mango", "papaya", "pineapple", "cherry")
(green, *tropic, red) = fruits
print(green)
print(tropic)
print(red)
```

```python
thistuple = ("apple", "banana", "cherry")
for x in thistuple:
  print(x)
```

```python
thistuple = ("apple", "banana", "cherry")
for i in range(len(thistuple)):
  print(thistuple[i])
```

```python
thistuple = ("apple", "banana", "cherry")
i = 0
while i < len(thistuple):
  print(thistuple[i])
  i = i + 1
```

```python
tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)
tuple3 = tuple1 + tuple2
print(tuple3)
```

```python
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2
print(mytuple)
```

```python
thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)
x = thistuple.index(8)   #Search for the first occurrence of the value
print(x)  
```

```python
thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)
x = thistuple.count(5)   #Return the number of times the value, appears in the tuple
print(x)
```
