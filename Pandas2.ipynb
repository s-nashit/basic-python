```python
df['Total'] = df['English']+df['Hindi']+df['Maths']+df['Science']+df['Social Science']
```

```python
df
```

```python
df['Percentage'] = (df['Total']/500)*100
```

```python
df
```

```python
df.loc[(df.Hindi == 75)]
```

```python
df.loc[(df['English'] >= 90) | (df['Maths'] >=90) & (df['Hindi']>= 90)]
```

```python
df['Grade'] = 'Pass'
```

```python
df
```

```python
# 85+ -> distinction
# 80-85 -> first
# 75-80 -> pass
# 70-75 -> fail

df.loc[df['Percentage'] >= 85, ['Grade']] = 'Distinction'
df.loc[(df['Percentage'] >=80) & (df['Percentage'] <85), ['Grade']] = 'First'
df.loc[(df['Percentage'] >=75) & (df['Percentage'] <80), ['Grade']] = 'Pass'
df.loc[(df['Percentage'] < 75), ['Grade']] = 'Fail'
df.loc[~(df['Percentage'] >= 0), ['Grade']] = 'Under Review'
```
