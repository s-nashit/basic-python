*Box Plot & Whisker Plot*

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,4))
x = [50, 60, 70, 80, 90]
plt.boxplot(x)
plt.savefig('box.png')
```

```python
plt.figure(figsize=(4,4))
x = [50, 60, 70, 80, 90]
plt.boxplot(x, notch = True)
plt.show()
```

```python
plt.figure(figsize=(4,2))
x = [50,55,0, 60, 70, 80, 90]
plt.boxplot(x, vert=False, widths= 0.3, labels= ["Student's Marks"], showmeans= True)
plt.show()
```
