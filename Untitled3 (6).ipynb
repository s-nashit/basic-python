```python
import numpy as np
import matplotlib.pyplot as plt

a = plt.figure()

ax = plt.axes(projection = '3d')

x = [0, 21, 33, 45, 57]
y = [0, 1, 3, 5, 7]
z = [10, 15, 23, 25, 27]

ax.plot3D(x, y, z, 'red')

ax.scatter3D(x, y, z)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([35, 25, 65])
b = ["Python", "Cloud", "BTML"]
c = ["pink", "red", "blue"]

plt.pie(a, labels = b, colors = c)
plt.legend()
plt.show()
```
