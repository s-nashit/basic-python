Merge

```python
df1 = pd.read_csv('details.csv')
df1
```

```python
df2 = pd.read_csv('marks.csv')
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on= 'Name')
```

```python
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'right')
```

Concat

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
a1 = pd.concat([df1, df2])
a1
```

```python
a2 = a1.groupby('Maths')
a2
```

```python
a2.get_group(93)
```

```python
for x in a2:
    print(x)
```

```python
for x, y in a2:
    print(x)
    print(y)
```
