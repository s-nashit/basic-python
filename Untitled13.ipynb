```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)

y_cos = np.cos(x)
y_sin = np.sin(x)

plt.plot(x, y_cos)
plt.plot(x, y_sin)

plt.xlabel('x- label')
plt.ylabel("amplitude")

plt.title('Electrical waves')
plt.legend(['Sine', 'Cosine'])
plt.show()
```
