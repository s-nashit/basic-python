```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Building models
mukesh = LinearRegression()
```

```python
#Training models
mukesh.fit(x_train, y_train)
```

```python
y_pred = mukesh.predict(x_test)
```
