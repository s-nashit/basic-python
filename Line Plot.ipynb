```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,2))
x = [1,2,3,4,5]
y = [78, 82, 86, 89, 92]
a = {'color':'red', 'size': 12, 'family': 'arial'}
plt.style.use('seaborn-v0_8-notebook')
#Title
plt.title('Mukesh Report', fontdict=a, loc = 'center')
# Labels
plt.xlabel('Class', fontsize = 8).set_color('red')
plt.ylabel('Marks').set_color('#c25478')
#figsize(width, height)
# dotted -> :; dashed-> -- ; solid -> -,dashdot -> -., none->''
plt.plot(x,y, color ='#c25478', marker = 'X')
plt.grid()
plt.savefig('patna.png')
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,3))

nsti = np.array(['Allahabad', 'Dehradun', 'Kanpur', 'Patna'])
strength = np.array([17, 21, 17, 10])
plt.bar(nsti, strength)
plt.show()
```
